```python
import os
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
from google.colab.patches import cv2_imshow
```

```python
# Caminho do banco de faces
db_path = "/content/drive/MyDrive/Projetos/datasets/facial_recognition/faces_db/dataset"

# Caminho da imagem de teste
test_img_path = "/content/drive/MyDrive/Projetos/datasets/facial_recognition/neymar_messi_e_cristiano.png"
```

```python
nome_real = {
    "messi": "LIONEL MESSI",
    "cr7": "CRISTIANO RONALDO",
    "neymar": "NEYMAR JR"
}
```

```python
# Carregar imagem
img = cv2.imread(test_img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

imgplot = plt.imshow(img_rgb)
```

```python
# Detectar múltiplos rostos
faces = DeepFace.extract_faces(
    img_path=test_img_path,
    detector_backend="mtcnn",
    enforce_detection=False,
    align=False
)
```

```python
for face in faces:
    area = face["facial_area"]
    x, y, w, h = area["x"], area["y"], area["w"], area["h"]

    # Normalizar imagem do rosto
    face_img = (face["face"] * 255).astype("uint8")

    # Salvar temporariamente
    temp_path = "/content/temp_face.png"
    cv2.imwrite(temp_path, cv2.cvtColor(face_img, cv2.COLOR_RGB2BGR))

    # Buscar no banco
    result = DeepFace.find(
        img_path=temp_path,
        db_path=db_path,
        enforce_detection=False,
        detector_backend="mtcnn",
        model_name="VGG-Face",
        distance_metric="cosine"
    )

    if not result[0].empty:
        best_match = result[0].iloc[0]
        file_name = os.path.splitext(os.path.basename(best_match["identity"]))[0]

        # Prefixo do arquivo (messi1 → messi)
        prefix = "".join([c for c in file_name if not c.isdigit()])

        # Nome bonito
        identity_clean = nome_real.get(prefix.lower(), prefix.upper())

        # Caixa verde com nome
        cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(img_rgb, identity_clean, (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    else:
        # Se não reconheceu → caixa vermelha
        cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (0, 0, 255), 2)
        cv2.putText(img_rgb, "Desconhecido", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

# Mostrar resultado
plt.figure(figsize=(12, 8))
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Reconhecimento: Messi, CR7 e Neymar")
plt.show()
```
